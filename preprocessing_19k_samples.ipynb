### Remove instructions with less than 100 tokens in response.

```python
# Remove instructions with less than 100 tokens in response.

from transformers import LongformerTokenizer

# Initialize the tokenizer (using tokenizer 'longformer-base-4096', because max squences length of longformer is 4096 which is greater than common tokenizers like bert-base-uncased and T5Tokenizer)
tokenizer = LongformerTokenizer.from_pretrained('allenai/longformer-base-4096')

# Token count filtering based on tokens, not words
def count_tokens(text):
    tokens = tokenizer.tokenize(text)
    return len(tokens)

# this creates a column at the end of dataset which contains the number of response tokens that particular row has.
flanV2_mix['response_token_length'] = flanV2_mix['targets'].apply(count_tokens)
```

```python
FLAN_2024_MIX = flanV2_mix[flanV2_mix['response_token_length'] >= 100]

# Reset index after filtering
FLAN_2024_MIX = FLAN_2024_MIX.reset_index(drop=True)
```

### Deduplication using cosine similarity (Threshold > 0.95)

```python
FLAN_2024_MIX['targets'].tolist()
```

```python
# Deduplication using cosine similarity

from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(FLAN_2024_MIX['targets'].tolist(), convert_to_tensor=True)
```

```python
embeddings
```

```python
embeddings = embeddings.cpu().numpy()
```

```python
from sentence_transformers import util

# Calculate cosine similarities
cosine_similarities = util.pytorch_cos_sim(embeddings, embeddings)

# Get indices of duplicates
duplicate_indices = set()
for i in range(len(cosine_similarities)):
    if i in duplicate_indices:
        continue
    for j in range(i + 1, len(cosine_similarities)):
        if cosine_similarities[i][j] > 0.95:
            duplicate_indices.add(j)
```

```python
len(list(duplicate_indices))
```

```python
# Remove duplicates
flanV2_deduplicated = FLAN_2024_MIX.drop(list(duplicate_indices))
```

### Plot a token distribution graph.

```python
# Plot token distribution
import matplotlib.pyplot as plt
%matplotlib inline

plt.hist(flanV2_deduplicated['response_token_length'], bins=50)
plt.xlabel('Token Count')
plt.ylabel('Frequency')
plt.title('Token Distribution in Responses')
plt.show()
```

- Now we drop the column `response_token_length` since we only used it for preprocessing the dataset.

```python
FlanV2_Final = flanV2_deduplicated.drop(columns=['response_token_length'])
```
